<b>Adding Labels and Title</b>

```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [5,7,4]

plt.plot(x, y,)
plt.xlabel("X Co-ordinates")
plt.ylabel("Y C0-ordinates")
plt.title('Interesting graph\nCheck it out')
plt.show()
```

<b>Adding legends</b>

```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [5,7,4]

x2 = [1, 2, 3]
y2 = [10, 14, 12]

#plt.plot(x, y, label='stock_price')

plt.plot(x, y, label='First Line')
plt.plot(x2, y2, label='Second Line')
plt.xlabel('Plot Number')
plt.ylabel('Important var')
plt.title('Interesting graph\nCheck it out')
plt.legend()
plt.show()
```

<b>Changing Colors and Line widths</b>

```python
import matplotlib.pyplot as plt

x1 = [1,2,3]
y1 = [5,7,4]

x2 = [1, 2, 3]
y2 = [10, 14, 12]

plt.figure(figsize=(15,6))
plt.plot(x1, y1, color="blue", linewidth=2.5, 
         linestyle="--", label="first line")
plt.plot(x2, y2, color="red", linewidth=2.5, 
         linestyle="-", label="second line")
plt.xlabel("plot number")
plt.ylabel("variables")
plt.legend()
plt.show()
```

Here we have changed the color, line_style and size of line for both x1, y1 and x2,y2 graph

<b>Setting Limits</b>

```python
import matplotlib.pyplot as plt

x1 = [1,2,3]
y1 = [5,7,4]

x2 = [1, 2, 3]
y2 = [10, 14, 12]

#toset pig size
plt.figure(figsize=(10,6))
plt.plot(x1, y1, color="blue", linewidth=2.5, 
         linestyle="-", label="first line")
plt.plot(x2, y2, color="red", linewidth=2.5, 
         linestyle="-", label="second line")
#setting limits
plt.xlim(1, 5)
plt.ylim(4, 18)
plt.xlabel("plot number")
plt.ylabel("variables")
plt.legend()
plt.show()


```

<h3>Bar Charts</h3>

With bar charts, each column represents a group defined by a categorical variable;

```python
x1 = [2,4,6,8,10]
y1 = [8,6,2,5,6]

x2 = [1,3,5,7,9]
y2 = [5,2,7,8,2]

plt.bar(x1,y1, label='Bar1', color='r')
plt.bar(x2, y2, label='Bar2', color='c')


plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting graph\nCheck it out')
plt.legend()
plt.show()
```

<h3>Pie Charts</h3>

```python
slices = [7, 2, 2, 13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']

plt.pie(slices, labels=activities,
       colors= cols,
       startangle = 90,
       shadow = True,
       explode = (0, 0.1, 0, 0),
       autopct = '%1.1f%%')

plt.title('Interesting Graph\nCheck it out')
plt.show()
```
